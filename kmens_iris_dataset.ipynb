```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
cols = ['sepal_length','sepal_width','petal_width','petal_length','class']
data = pd.read_csv('C:\\Users\\UTSAV\\Desktop\\MY_Practice\\prac\\iris.data',names= cols,header=None)
data.head(5)
```

```python
x = data.iloc[:,:-1]
```

```python
var = []
for k in range(1,11):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(x)
    var.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11),var)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(x)
kmeans.labels_ 
kmeans.fit_transform(x)
```

```python
c = kmeans.cluster_centers_
```

```python
a = [5.9,3.0,5.1,1.8]
pred = kmeans.predict([a])
pred
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(16,9))
ax = Axes3D(fig)
ax.scatter(x.iloc[:, 0],x.iloc[:, 1],x.iloc[:, 2])
ax.scatter(c[:, 0], c[:, 1], c[:, 2], marker='*', c='red')
```
